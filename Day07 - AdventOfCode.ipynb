```python
with open('Input07') as f:
    data = f.read()
```

```python
import regex as re
from collections import defaultdict
```

```python
terminal = data.split('\n')
```

```python
terminal
```

```python
outermost_directory = re.compile(r"\$ cd /")
in_directory = re.compile(r"\$ cd (\w+)")
out_directory = re.compile(r"\$ cd ..")
ls = re.compile(r"\$ ls")
directory = re.compile(r"dir (\w+)")
size = re.compile(r"(\d+) (\w+)") # va surement pas marcher pck y a pas que des lettres mais aussi des points
```

```python
directories = defaultdict(dict)
```

```python
current_directory = dict()
previous_directory = dict()

for ligne in terminal:
    
    match_in = in_directory.match(ligne)
    match_dir = directory.match(ligne)
    match_size = size.match(ligne)

    if outermost_directory.match(ligne):     
        directories['outermost']
        current_directory = directories
        
    if in_directory.match(ligne): 
        previous_directory = current_directory 
        previous_directory[match_in.groups()] = dict() 
        
    if out_directory.match(ligne):
        current_directory = previous_directory
    
    if match_dir:
        current_directory[match_dir.groups()[0]] = dict()
            
    if match_size:
        current_directory[match_size.groups()[1]] = int(match_size.groups()[0])
```
